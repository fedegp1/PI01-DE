```python
import pandas as pd
```

```python
nombre_archivo = 'sucursal.csv'
# este nombre tiene que tomarse de la ejecución del script; para desarrollarlo lo hardcodeamos
extension = nombre_archivo.split('.')[1]
# guarda la extension del archivo
ruta = './Datasets/'
# harcodeamos ruta relativa

```

```python
dfsucursal = pd.read_csv(str(ruta)+str(nombre_archivo))
```

```python
# observamos valores de latitud y longitud para argentina. No encontramos incongrüencias.
# los valores extremos (maximos y minimos) de ambos parámetros estan comprendidos entre los posibles.
# lat -21 ~ -56 // lng -53 ~ -74
```

```python
def sacar_espacios(df):
    '''
    Quitar espacios vacíos de de ambos lados de las cols con strings.
    '''
    sacar_espacios = lambda x: x.strip() if isinstance(x, str) else x
    return df.applymap(sacar_espacios)
sacar_espacios(dfsucursal)
```

```python
def minuscula(df):
    '''
    convierte a minusculas todos los textos del dataframe
    '''
    minuscula = lambda x: x.lower() if isinstance(x, str) else x
    return df.applymap(minuscula)
dfsucursal = minuscula(dfsucursal)

```

```python
def titular(df):
    '''
    Convierte a mayúscula la 1era letra de cada palabra en todas las cols.
    '''
    titular = lambda x: x.title() if isinstance(x, str) else x
    return df.applymap(titular)
dfsucursal = titular(dfsucursal)
```

```python
# mantengo en mayúsculas la columna 'provincia'
dfsucursal['provincia']=dfsucursal['provincia'].str.upper()
```

```python
dfsucursal.describe(include=[object])
#importante chequear que no haya Ids duplicados. Se observa que countid y uniques son iguales. entonces no hay duplicidad de ids.
```

```python
dfsucursal.isnull().sum()
#chequeo nulos y observo que no hay valores nulos en toda la tabla
```

```python
# debo chequear también que no haya errores en la columna Id, en cuanto que este id se conforma al concatenar la columna banderaID y comerciorazon social.
# descompongo la col id, en los 3 valores que la conforman para luego chequear que indivualmente coincidan con las columnas correspondientes.

dfsucursal_check = dfsucursal['id'].str.split('-', n=2, expand=True)
```

```python
check1 = dfsucursal_check[0] == dfsucursal_check['comercioId'] 
check1.describe()
#observo que es True para toda la columna. no hay incongrüencias
```

```python
check1 = dfsucursal_check[1] == dfsucursal_check['banderaId'] 
check1.describe()
#observo que es True para toda la columna. no hay incongrüencias
```
